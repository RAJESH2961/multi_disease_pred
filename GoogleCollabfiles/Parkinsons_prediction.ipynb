# Data PreProcessing
seperating the feactures & Target




```python
X = parkinsons_data.drop(columns=['Diagnosis','PatientID', 'Age', 'EducationLevel','DoctorInCharge'], axis=1)
Y = parkinsons_data['Diagnosis']
```

```python
# Here X is an feacture
# Y is an Targeted Variable
# In feacure no need to give an name of the patient there is no use of that
print(Y)
```

Splitting the Data into Training and Testing


```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Choosing the Model and Training the splitted Data
Model --> SVM(Support Vector Machine)

```python
model = svm.SVC(kernel='linear')
model.fit(X_train, Y_train)
```

Accuracy Score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
#Accuracy Score on testing Data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of test data : ', test_data_accuracy)
```
